```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 


df= pd.read_csv('C:\\Users\\pc\\Desktop\\top skill required.csv')
# Analyze the data by grouping based on job titles, skills, and salaries
# Find the top-paying jobs and the most frequent skills

# Count the frequency of each skill
skill_counts = df['skills'].value_counts().head(10)

# Plot a bar graph showing the top 10 most required skills
plt.figure(figsize=(10, 6))
sns.barplot(x=skill_counts.values, y=skill_counts.index, palette="viridis")
plt.title('Top 10 Most Required Skills in Data Analyst Roles')
plt.xlabel('Number of Job Postings')
plt.ylabel('Skills')
plt.show()
```

```python
# Data for the pie chart
skills = ['SQL', 'Excel', 'Python', 'Tableau', 'Power BI']
job_demand_count = [110380, 75348, 68409, 56569, 45482]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(job_demand_count, labels=skills, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.title('Job Demand for Different Skills in Data Analyst Roles')
plt.show()
```
